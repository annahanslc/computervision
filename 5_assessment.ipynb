```python
# Open image

import cv2
import numpy as np
import matplotlib.pyplot as plt

img = plt.imread('DATA/dog_backpack.jpg')
```

```python
plt.imshow(img)
```

```python
img = cv2.imread('DATA/dog_backpack.jpg')
```

```python
plt.imshow(img)
```

```python
img_cvt = cv2.cvtColor(img,cv2.COLOR_RGB2BGR)
```

```python
plt.imshow(img_cvt)
```

```python
img_flip = cv2.flip(img_cvt,0)
plt.imshow(img_flip)
```

```python
cv2.rectangle(img_cvt,(200,400),(600,700),(255,0,0),5)
plt.imshow(img_cvt)
```

```python
vertices = np.array([[400,400],[600,700], [200,700]],dtype=np.int32)
pts = vertices.reshape((-1,1,2))

cv2.polylines(img_cvt,[pts],isClosed=True,color=(0,0,255),thickness=5)
cv2.fillPoly(img_cvt,[pts],(0,0,255))
plt.imshow(img_cvt)
```
