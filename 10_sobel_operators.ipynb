```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def display_img(img):
  fig = plt.figure(figsize=(12,10))
  ax = fig.subplots(1)
  ax.imshow(img,cmap='gray')
```

```python
img = cv2.imread('DATA/sudoku.jpg',0) # 0 to read in grayscale
```

```python
display_img(img)
```

```python
# calculate an x gradient sobel

sobelx = cv2.Sobel(img,cv2.CV_64F,1,0,ksize=5)
```

```python
display_img(sobelx)
```

```python
# calculate an y gradient sobel

sobely = cv2.Sobel(img,cv2.CV_64F,0,1,ksize=5)
```

```python
display_img(sobely)
```

```python
# Laplace operator
# in 2 dimensions, the laplace operator is the 2nd derivative of x plus the 2nd derivative of y
# edge detection

laplacian = cv2.Laplacian(img,cv2.CV_64F)
```

```python
display_img(laplacian)
```

```python
# blend the sobel x and sobel y

blended = cv2.addWeighted(src1=sobelx,alpha=0.5,src2=sobely,beta=0.5,gamma=0)
```

```python
display_img(blended)
```

```python
ret,th1 = cv2.threshold(img,100,255,cv2.THRESH_BINARY)
```

```python
display_img(th1)
```

```python
kernel = np.ones((4,4),np.uint8)
```

```python
gradient = cv2.morphologyEx(blended,cv2.MORPH_GRADIENT,kernel)
```

```python
display_img(gradient)
```
